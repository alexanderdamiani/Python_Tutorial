## DataFrame attributes and methods

```python
import pandas as pd 
import numpy as np

# just importing this will automatically format the graphs to a more appealing format
# comment this out and generate the plots for the dataframe below to compare to with/without seaborn imported
import seaborn as sns

# this line makes graphs show up in the jupyter notebook, not necessary outside of it
%matplotlib inline
```

```python
# create a DataFrame with dummy data
df = pd.DataFrame(np.random.sample((10,4)))
df
```

```python
# assign columns and index of DataFrame
df.columns = ['A', 'B', 'C', 'D']
df.index = range(1,11)
df
```

```python
# head and tail methods will give you a snapshot of the top and bottom of the DataFrame
df.head()
```

```python
df.tail()
```

```python
# Transpose your data
df.T
```

```python
# sort the DataFrame by a specific column
# inplace=True specifies to change the DataFrame, don't have to do df = df.sort_values...
df.sort_values(by='A', inplace=True)
# can also do multiple columns as df.sort_values(by=['A', 'B'], inplace=True)
df.head()
```

```python
# can reset the index so it doesn't appear out of order above
# again inplace=True prevents us from having to assign df = df.reset_index...
# drop=True says to drop the original index, otherwise is added as a column - columns would be ['index', 'A', 'B', 'C', 'D']
df.reset_index(inplace=True, drop=True)
df.head()
```

```python
# Selecting data by column
df['A'][:2]
```

```python
# Selecting data by index
df.iloc[:2]['A']
```

```python
# Working with missing data
df['A'] = [np.nan] + list(range(8)) + [np.nan]
df.head()
```

```python
# fills all NaN values with the first argument you specify across the whole DataFrame
# can specify to only focus on a specific column by doing df['A'].fillna(100, inplace=True)
df.fillna(1, inplace=True)
df.head()
```

```python
# plotting DataFrames
# cumsum function does cummulative sum --> [1,2,3] would become [1, 3, 6]
# drop column 'A' because values are higher than other columns, just for visual purposes (not necessary)
df.drop('A', axis=1, inplace=True)
df = df.cumsum()
df.plot()
```

## Data Analysis with pandas DataFrames

We will be working with data on Pokemon for this example.
We will follow a 3 step process:
- (1) clean the data
- (2) visualize the data
- (3) analyze the results

```python
df = pd.read_csv('Pokemon.csv')
df.head(20)
```

### Clean the Data

Looking at the first 20 rows we can make some observations:
- (1) the column # is unnecessary so we can remove it
- (2) in the Name column every name with Mega is preceded by a duplicate of the word after it. We can remove everything before Mega.
- (3) the Type 2 column has many NaN values. Logical conclusion is to fill the Type 2 column with the Type 1 value for the associated row

```python
# axis=1 specifies to drop columns and not rows
df.drop('#', axis=1, inplace=True)
```

```python
# we first set the type of the column to str with df['Name'].str
# then we apply a function of the str type in python, replace()
# replace() takes two arguments: (1) the string to search for, (2) the string to replace the first string with if its found
df['Name'] = df['Name'].str.replace('.*Mega', 'Mega')
```

```python
# fillna of Type 2 with Type 1
df['Type 2'] = df['Type 2'].fillna(df['Type 1'])
```

```python
df.head(20)
```

```python
# check to make sure no other NaN values in our DataFrame
df.isnull().sum()
```

```python
# Show a quick summary of the DataFrame to ensure all values look probable
df.describe()
```

### Visualizations of the Data

```python
# We're going to perform analysis based on the types, so first see what types are available
df['Type 1'].value_counts()
```

```python
# only plot the top ten values '[:10]', since pie chart gets too crowded
# autopct='%1.1f%%' is an argument passed to display percentages inside of each pie slice
df['Type 1'].value_counts()[:10].plot(kind='pie', autopct='%1.1f%%')
```

We see psychic and grass each compose a large portion of the number of types. So we'll perform our analysis on them.

```python
# Filter out rows where Type 1 or Type 2 is Psychic, or Grass
df_psychic = df[(df['Type 1'] == 'Psychic') | (df['Type 2'] == 'Psychic')]
df_grass = df[(df['Type 1'] == 'Grass') | (df['Type 2'] == 'Grass')]
```

```python
# plot the two selected types on the same graph
# assign the first plot to an object 'ax' so we can pass it as a parameter to the second DataFrame plot, 
# so they'll plot on the same graph

# label argument is necessary for the ax.legend() to know what labels to assign for each dataset
ax = df_psychic.plot(kind='scatter', x='Attack', y='Defense', color='purple', label='Psychic')
df_grass.plot(kind='scatter', x='Attack', y='Defense', color='green', label='Grass', ax=ax)
ax.legend()
```

### Data Analysis

We can draw a few conclusions from the plot above.
- psychic type pokemon have a larger range of possible attack and defense values
- psychic type pokemon have a larger distribution of attack and defense combinations than grass types
- even with the previous two points, we see a lot of overlap between the psychic and grass type pokemon attack and defense pairings
- if we needed to chose a pokemon purely based on this information, the answer isn't clear cut
    - psychic pokemon have higher possible attack and defense combinations, but also more lower combinations (especially high attack and low defense)
    - whether we are prioritizing attack (likely chose psychic), or defense (likely chose grass) would decide our choice of pokemon
